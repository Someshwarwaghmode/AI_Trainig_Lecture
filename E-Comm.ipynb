```python
plt.pie(df['Gender'].value_counts(),labels = {"F","M"} , autopct = "%1.1f%%",startangle = 90)
plt.title("Gender Distribution")
```

```python
fig ,  ax= plt.subplots(1,3 , figsize=(15,5))
sns.histplot(df['Weight_in_gms'] , ax=ax[0],kde=True).set_title("Wighr Distribution")
sns.countplot(x='Product_importance',data=df , ax=ax[1]).set_title("Product_importance")
sns.histplot(df['Cost_of_the_Product'] , ax=ax[2],kde=True).set_title('Cost_of_the_Product')
```

```python
#Logistic 
fig ,  ax= plt.subplots(1,3 , figsize=(15,5))
sns.countplot(x='Warehouse_block',data=df , ax=ax[0]).set_title("Warehouse_block")
sns.countplot(x='Mode_of_Shipment',data=df , ax=ax[1]).set_title("Mode_of_Shipment")
sns.countplot(x='Reached.on.Time_Y.N',data=df , ax=ax[2]).set_title("Reached.on.Time")
```
